# Var and SD

```python
Online_Retail=pd.read_csv("data/Online Retail.csv",encoding = "ISO-8859-1")
```

```python
Online_Retail.info()
```

```python
#var and sd UnitPrice
var_UnitPrice=Online_Retail['UnitPrice'].var()
print(var_UnitPrice)

std_UnitPrice=Online_Retail['UnitPrice'].std()
print(std_UnitPrice)
```

```python
var_qty=Online_Retail['Quantity'].var()
print(var_qty)

std_qty=Online_Retail['Quantity'].std()
print(std_qty)
```
